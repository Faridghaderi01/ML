Generalization
=

E[(y - f{x} ** 2] = Bias ** 2 + Variance + noise
-

____________________________

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = np.array([[50], [60], [70], [80], [90], [100]])
y = np.array([[101],[123],[141],[162],[181],[202]]) 

#______________________________________
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
#__________________________________________
n = np.array([[500], [250]])
y_pred = model.predict(n)
print(y_pred)
```

```python
#Bias
bias = np.mean(y_pred) - np.mean(y_test)
#variance
variance = np.var(y_pred)
#expected_error
noise = 0.1
expected_error = bias**2 + variance + noise

print("بایاس:", bias)
print("واریانس:", variance)
print("خطای غیرقابل کاهش:", noise)
print("امید ریاضی (E[(y - f(x))^2]):", expected_error)
```

این مقادیر بایاس و واریانس و نتیجه انها یعنی امید ریاضی عدد بزرگی هستند و هر چقدر عدد کوچکتر باشد نشان دهنده عملکرد درست مدل است 
در نتیجه برای کم کردن بایاس میتوان تعداد پارامتر ها را بیشتر کرد 
و برای کم کردن واریانس میتوان داده ها بیشتری به مدل داد  

```python
plt.scatter(X, y, color='blue', label='real data')
plt.scatter(X_test, y_pred, color='red', label='predicted')
plt.plot(X, model.predict(X), color='green', label='linear regrission')
plt.xlabel(' (size )')
plt.ylabel('price')
plt.title('regrission model')
plt.legend()
plt.show()
```
