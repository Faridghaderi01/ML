SVD
---


تجزیه مقدار منفرد یک تکنیک ریاضی است که به شما اجازه می‌دهد یک ماتریس را به سه ماتریس دیگر تجزیه کنید


این تکنیک می‌تواند برای تجزیه و تحلیل داده‌ها، کاهش ابعاد، و استخراج ویژگی‌ها در یادگیری ماشین و آمار مفید باشد

```python
import numpy as np


x = np.array([[1, 2], [3, 4], [5, 6]])

# محاسبه مجموع مربعات هر ستون
sum_squares = np.sum(x * x, axis=0)

# ساخت ماتریس جدید با استفاده از مجموع مربعات و x
A = np.multiply(sum_squares, x)

# ضرب ماتریس A با ترانهاده x
A = np.dot(A, x.T)

# انجام SVD
U, S, Vt = np.linalg.svd(A)

```
