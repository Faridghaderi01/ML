```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, ConfusionMatrixDisplay
```

Euclidean distance
-

```python
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix - Euclidean Distance")
plt.show()

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', label='Train Data')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, marker='x', label='Predictions')
plt.title("KNN Classification with Euclidean Distance")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

_______________________________________________

Minkowski
-

```python
knn_minkowski = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=3)  
knn_minkowski.fit(X_train, y_train)
y_pred_minkowski = knn_minkowski.predict(X_test)


accuracy_minkowski = accuracy_score(y_test, y_pred_minkowski)
precision_minkowski = precision_score(y_test, y_pred_minkowski)
recall_minkowski = recall_score(y_test, y_pred_minkowski)
f1_minkowski = f1_score(y_test, y_pred_minkowski)

print(f"Accuracy (Minkowski): {accuracy_minkowski:.2f}")
print(f"Precision (Minkowski): {precision_minkowski:.2f}")
print(f"Recall (Minkowski): {recall_minkowski:.2f}")
print(f"F1 Score (Minkowski): {f1_minkowski:.2f}")

cm_minkowski = confusion_matrix(y_test, y_pred_minkowski)
disp_minkowski = ConfusionMatrixDisplay(confusion_matrix=cm_minkowski)
disp_minkowski.plot()
plt.title("Confusion Matrix - Minkowski Distance")
plt.show()

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, marker='o', label='Train Data')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred_minkowski, marker='x', label='Predictions')
plt.title("KNN Classification with Minkowski Distance")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Cosine distance regression
-

```python
from sklearn.neighbors import KNeighborsRegressor


X_regression, y_regression = make_classification(n_samples=100, n_features=2, n_classes=1, n_informative=2, n_redundant=0, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_regression, y_regression, test_size=0.3, random_state=42)


knn_cosine = KNeighborsRegressor(n_neighbors=5, metric='cosine')
knn_cosine.fit(X_train_reg, y_train_reg)
y_pred_reg = knn_cosine.predict(X_test_reg)

r2_score = knn_cosine.score(X_test_reg, y_test_reg)
print(f"R^2 Score (Cosine Distance Regression): {r2_score:.2f}")

plt.scatter(X_test_reg[:, 0], y_test_reg, color='blue', label='True Values')
plt.scatter(X_test_reg[:, 0], y_pred_reg, color='red', label='Predictions')
plt.title("KNN Regression with Cosine Distance")
plt.xlabel("Feature 1")
plt.ylabel("Target Value")
plt.legend()
plt.show()
```
